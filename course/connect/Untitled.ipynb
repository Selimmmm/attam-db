```python
import psycopg2

USER = "postgres"
PASSWORD = ""

conn = psycopg2.connect(
    user=USER, 
    password=PASSWORD , 
    host="localhost", 
    port="5432"
)


cur = conn.cursor()
cur.execute("SELECT version();")
db_version = cur.fetchone()
print(db_version)
```
